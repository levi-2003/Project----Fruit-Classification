```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/major project/Book1.csv")
df
```

```python
df.drop(columns='fruit_label')
df.duplicated().sum()
df.drop(columns='fruit_label')
df['fruit']=df['fruit_name']+' '+df['fruit_subtype']
df
```

```python
df=df.drop(columns=['fruit_label',"fruit_name",'fruit_subtype'])
```

```python
df.groupby('fruit').count()
df
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=['fruit'])
y=df['fruit']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=21)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train=KNeighborsClassifier(n_neighbors=3)
train.fit(X_train,y_train)
```

```python
y_pred=train.predict(X_test)
```

```python
y_test=np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train2=RandomForestClassifier(n_estimators=100)
train2.fit(X_train,y_train)
```

```python
y_pred2=train2.predict(X_test)
```

```python
y_test=np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred2)
```

```python
import joblib
joblib.dump(train2,'fruits')
```
